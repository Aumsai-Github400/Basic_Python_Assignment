**5. How many global scopes are there in a Python program? How many local scopes?**

**Answer:**

- There are four scope in python programming such as Built-in scope, Global scope, Non-Local or Enclosing scope and at the last Local scope.
- Whatever variable defined outside a function is in Global scope of that function, and whatever variable defined inside function is in local scope of that function.
- By default there is one global scope and local scope is whenever function is created and called.
- To find how many global scope is there we can use function called globals().
- To find how many local scope is there we can use function called locals().

**14. What is the purpose of the try clause? What is the purpose of the except clause?**

**Answer:**

- In python error can be categories into two errors I.e.  syntax error and exception error. Errors tend to stop execution of program. To handle such problem in python there are two blocks named as Try block and handle block.


*    **Try Block**

    Try block use to check error in code, code that can potentially cause error will go inside try code. If error comes or happen try block will skip code and go for next code to execute


*    **Except Block**

    Except black use to handle error, whatever error cause into try block can be handle with the help of except block. And will execute rest code without exiting from code.
