**17. How do you distinguish between copy.copy() and copy.deepcopy()?**

**Answer:**

    In Python, we use = operator to create a copy of an object but it will create just refence to original object. (id are same for original one and new one)
    We can copy original object into new object by using copy module, there are two ways to copy object.


- Shallow copy:


    Shallow copy creates new object which store reference of nested object from original object. Hence if old object gets updated or modified then results can appear in new object as well.


- Deep copy:
    
    
    Deep copy creates new object and copy nested object too. Hence if old copy gets updated or modified results appeared in new one.
